```python
import numpy as np
import gmms
```

# Example: 2003 Tokachi Earthquake

## Earthquake parameters

```python
M       = 8.29
hyp_lat = 41.7796
hyp_lon = 144.0786
hyp_Z   = 25.0101
```

## Fault model

```python
ULC_lat = 42.1796
ULC_lon = 144.8114
fZtor = 23.3
flength = 120.
fwidth  = 100.
fstrike = 230.
fdip    = 20.
frake   = 90.
fins = 0
fint = 1
```

## Site parameters

```python
site_lat  = 40.9255
site_lon  = 140.9527
site_Vs30 = 289.219
```

## Compute fault parameters

```python
[URC_lat,URC_lon] = gmms.faulttools.get_fault_URC(ULC_lat,ULC_lon,flength,fstrike)
[fnm,frv,fmech]   = gmms.faulttools.get_faulting_style(frake)
```

## Compute site parameters

```python
[_,R_rup,_] = gmms.distancetools.get_distances(site_lat,site_lon,ULC_lat,ULC_lon,URC_lat,URC_lon,fwidth,flength,fdip,fstrike,fZtor)
```

```python
print('Rrup = {:.2f} km'.format(R_rup[0]))
```

## Compute ground motion intensity measures

### Ia

```python
[mu,phi,tau] = gmms.FoulserPiggottGoda2015.FoulserPiggottGoda2015(0,M,hyp_lat,hyp_lon,hyp_Z,R_rup,site_Vs30,fins,fint,fnm,frv)
print('Ia = {:.2f} m/s'.format(np.exp(mu[0])))
```

### CAV

```python
[mu,phi,tau] = gmms.FoulserPiggottGoda2015.FoulserPiggottGoda2015(1,M,hyp_lat,hyp_lon,hyp_Z,R_rup,site_Vs30,fins,fint,fnm,frv)
print('CAV = {:.2f} m/s'.format(np.exp(mu[0])))
```
